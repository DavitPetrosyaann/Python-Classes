# Why Use Built-in Objects?

Python’s built-in objects (like lists, dictionaries, tuples, and sets) simplify programming by providing ready-to-use, efficient data structures. They remove the need to manually create and manage objects as in lower-level languages.

* Easier programming: Built-in objects let developers quickly represent data and solve problems without writing extra code.

* Foundation for extensions: Custom classes and data structures are often built on top of built-in types (e.g., a stack can be made using a list).

* High performance: Built-in objects are optimized and implemented in C, making them faster and more efficient than user-defined structures.

* Standard and portable: They are part of Python’s core, ensuring consistency and compatibility across all environments.

* Custom data types: When needed, Python allows you to define your own object data types using classes, extending or customizing the behavior of built-in objects.


In short: Built-in objects make Python programming easier, faster, and more reliable, while still allowing the creation of custom data types when special functionality is required.

## Summary

| Level                                  | Responsibility                                   |
| -------------------------------------- | ------------------------------------------------ |
| Hardware                               | Stores and manipulates 0s and 1s                 |
| Encoding standard (e.g., ASCII, UTF-8) | Defines what bit patterns mean as characters     |
| Programming language                   | Assigns meaning via data types                   |
| Application / file format              | Uses conventions to decide how to interpret bits |


```
# The number 65 in decimal
num = 65

# Convert the number to binary
binary = format(num, '08b')
print("Binary representation:", binary)   # 01000001

# Interpret it as an ASCII character
char = chr(num)
print("As ASCII character:", char)        # A

# Now go the other way around
char2 = 'A'

# Get the numeric ASCII code for 'A'
num2 = ord(char2)   # reverse of chr()
print("Numeric value of 'A':", num2)      # 65

# Verify both share the same binary pattern
print("Binary of 'A':", format(num2, '08b'))  # 01000001
```

## In contrast — C-style strings do this

In low-level C:

    char s[] = "Hello";


This is stored in memory as:

    'H' 'e' 'l' 'l' 'o' '\0'


That \0 (null terminator) marks the end — it’s an extra byte (0x00) so that the program can tell where the string ends, since C doesn’t store the length separately.

## But Python strings are objects

When you create:

    s = "Hello"


Python stores something like this internally (simplified):

| Field       | Meaning                                  |
| ----------- | ---------------------------------------- |
| `ob_refcnt` | Reference count                          |
| `ob_type`   | Type pointer (`str`)                     |
| `length`    | Number of characters (e.g. 5)            |
| `data`      | Actual characters: `'H' 'e' 'l' 'l' 'o'` |

So when Python needs to iterate or slice, it just reads the length — no null terminator needed.


### All stored in heap memory

All Python objects (except a few internal caches like small ints) are allocated on the heap.
The Python memory manager handles these allocations, not the operating system directly.

* Python’s heap = managed memory pool for PyObjects

* Garbage collector = frees objects when ob_refcnt drops to 0

## Changing lists in place

Lists are **mutable**, so they support operations that change a list object in place (by overwriting its former value).

```
my_list = [1, 3, 5, 7, 9]

print('Original list: ', my_list, end="\n\n")
my_list[2] = 8
print("Index assignment: ", my_list, end="\n\n")

my_list[:2] = ['one', 'three']
print('Slice assignment: ', my_list, end="\n\n")

my_list[2:3] = ['four', 'five']
print('Inserting an additional element during the replacement: ', my_list, end="\n\n")

my_list[1:1] = ['two']
print('Inserting without replacement: ', my_list, end="\n\n")

my_list[5:7] = []
print('Deletion (inserting nothing): ', my_list, end="\n\n")

del my_list[0]
print('Other way to delete an item: ', my_list, end="\n\n")

del my_list[:2]
print('Delete a slice: ', my_list, end="\n\n")
```

**Making copies of lists:** Making copies of lists is a little tricky due to the way Python handles lists.
Say we have a list `L` and we want to make a copy of the list and call it `M`. The expression `M = L` will
not work.

```
L = [1,2,3]
z = L
L[0] = 9
print('L is now:', L, ' Copy:', z)
```

We can see that the code did not work as we might have expected. When we changed `L`, the `copy` got changed as well. We can use the following code to actually copy (**shallow copy**) a list.

```
a = [7, 8]
L = [1,2,3]
# copy = L[:] # this line copies L
copy = L.copy()
L[0] = 9
print('L is now:', L, ' Copy:', copy)
```

```
a = [7, 8]
L = [1,2,3, a]
copy = L.copy()
L[-1][0] = 9
print('L is now:', L, ' Copy:', copy)
```

```
import copy
a = [7, 8]
L = [1,2,3, a]
copy = copy.deepcopy(L)
L[-1][0] = 0
L[0] = 9
print('L is now:', L, ' Copy:', copy)
print(id(L))
print(id(copy))
```

* Making a shallow copy of an object won't clone child objects. Therefore, the copy is not fully independent of the original.
* A deep copy of an object will recursively clone child objects. The clone is fully independent of the original, but creating a deep copy is slower.

The reason behind this behaviour is **referencing**.

Remember that everything in Python is an object. This includes **numbers**, **strings**, and **lists**. When we do a simple variable assignment, like `x=487`, what actually happens is Python creates an integer
object with the value `487`, and the variable `x` acts as a **reference to that object**. If we come along and declare `y=487`, then `y` also points to that same memory location.

When we set `L=[1,2,3]`, we create a list object `[1,2,3]` and a reference, `L`, to it. When we say
`copy=L`, we are making another reference to the object `[1,2,3]`. When we do `L[0]=9`, **because
lists are mutable**, the list `[1,2,3]` is changed in place to `[9,2,3]`. No new object is created.
The list `[1,2,3]` is now gone, and since copy is still pointing to the same location, it’s value is
`[9,2,3]`.

On the other hand, if we instead use `copy=L[:]`, we are actually **creating a new list object** somewhere
else in memory so that there are two copies of `[1,2,3]` in memory. Then when we do
`L[0]=9`, we are only changing the thing that `L` points to, and copy still points to `[1,2,3]`.

```
# sorting a list (by default ascending (increasing) order )
new_list.sort()
new_list  # upper case letters come first
```

```
# sorting in descending order
new_list.sort(reverse=True)
new_list
```

```
new_list = ['Ab', 'a', 'g', 'b']
```

```
# we can ignore the case type by considering the items as lower case
new_list.sort()
new_list
```

* `reverse()` - reverses the list

```
new_list = ['We', 'need', 'a', 'list']
# reverse the list in place
new_list.reverse()
print(new_list)
```

**Important note:** There is a big difference between `list methods` and `string methods`:

* String methods do not change the original string

* List methods do change the original list.

For example, to sort a list `L`, just use `L.sort()` and not `L = L.sort()`.

## Random elements from a list

## Split, Join strings

The `split()` method returns a list of words of a string. The method assumes that words are separated
by whitespace or other characters. Here is an example:

```
s = 'Here are some words.'
print(s.split(), end = '\n\n')

s = 'Here - are - some - words.'
print(s.split(' - '), end = '\n\n')
```

The `join()` method is the opposite of `split()`. It is a string method that takes a `list of strings` and **joins them together** into a single string. Here are some examples:

```
L = ['a', 'b', 'c']
print(''.join(L), end = '\n\n')
print(' '.join(L), end = '\n\n')
print('-|-'.join(L), end = '\n\n')
print(' '.join(L), end = '\n\n')
```
