#Task 1. We have the solution of the system of two linear equations, now we have to optizime the solution.

```python
%%timeit
lst = []

for x in range(-50, 51):
  for y in range(-50, 51):

    #print(x, y)
    if 2*x + 3*y == 4 and x - y ==7:
      lst.append((x, y))

lst
```

# Task 3. Leetcode problem `Two Sum`

```python
lst =  [1, 5, 6, 8*9, 7/9, -7, 45, 456]

def find_neg_num(lst_numbers):
  index = 0

  while index < len(lst_numbers):
    if lst_numbers[index] < 0:
      return lst_numbers[index]
    index += 1
  else:
    return 'No negative number'



find_neg_num(lst)
```
