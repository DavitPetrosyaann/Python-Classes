#Argument Unpacking

As we have already know the single asterisk operator * can be used on any iterable that Python provides, while the double asterisk operator ** can only be used on dictionaries. We can use this unpacking method in functions.

##*args

If you do not know how many arguments that will be passed into your function, add an * before the parameter name in the function definition.

This way the function will receive a tuple of arguments, and can access the items accordingly.

````python
def multiply(*args):
    result = 1
    print(type(args))
    for arg in args:
        result *= arg

    return result

multiply(1, 2, 3, 4, 5, 56, 7)
````

We can have fixed argument(s)

````python
def multiply(arg1, *args):
    result = arg1
    print("First argument :", arg1)
    for arg in args:
        result *= arg
    return result

multiply(21, 15)
````

##**kwargs
We can also send arguments with the **key = value** syntax.

````python
def concatenate(**kwargs):
    result = ""
    print(type(kwargs), end = '\n\n')
    print(kwargs, end = '\n\n')

    # Iterating over the Python kwargs dictionary
    for arg in kwargs.values():
        result += arg + ' '

    return result

print(concatenate(a="I", b="am", c="Rafo", d="!"))
````

We can use keys of **kwargs

````python
def concatenate(**kwargs):
    result = ""
    # Iterating over the keys of the Python kwargs dictionary
    for arg in kwargs:  # kwargs.keys()
        result += arg
    return result

print(concatenate(a="I", b="am", c="Rafo", d="!"))
````

# Recursion

In Python, we know that a function can call other functions. It is even possible for the function to call itself. These types of construct are termed as recursive functions.




image.png

image.png

````python
def factorial(x):
    """This is a recursive function
    to find the factorial of a given integer
    """

    if x == 1:
        return 1
    return x * factorial(x-1)


num = 3
print("The factorial of", num, "is", factorial(num))
````

```
factorial(3)          # 1st call with 3

3 * factorial(2)      # 2nd call with 2

3 * 2 * factorial(1)  # 3rd call with 1

3 * 2 * 1             # return from 3rd call as number=1

3 * 2                 # return from 2nd call

6                     # return from 1st call
```

## Maintaining State

When dealing with recursive functions,  each recursive call has its own execution context, so to maintain state during recursion you have to either:

* Thread the state through each recursive call so that the current state is part of the current call's execution context
* Keep the state in global scope

Let's calculate 1 + 2 + 3 ⋅⋅⋅⋅ + 10 using recursion. The state that we have to maintain is (current number we are adding, accumulated sum till now)

````python
def sum_recursive(current_number, accumulated_sum):
    # Base case
    # Return the final state
    if current_number == 11:
        return accumulated_sum

    # Recursive case
    # Thread the state through the recursive call
    else:
        return sum_recursive(current_number + 1,
                             accumulated_sum + current_number)
````

## Advantages of Recursion

* Recursive functions make the code look clean and elegant.
* A complex task can be broken down into simpler sub-problems using recursion.

## Disadvantages of Recursion

* Sometimes the logic behind recursion is hard to follow through.
* Recursive calls are expensive (inefficient) as they take up a lot of memory and time.
* Recursive functions are hard to debug.

Remember Fibonacci numbers

1, 1, 2, 3, 5, 8, 13, 21 ...

$f_1=f_2=1$

$f_n=f_{n-1}+f_{n-2}, \, n=3,...$


````python
def fibonacci(n):
  if n <= 2:
    return 1
  return fibonacci(n-1) + fibonacci(n-2)

fibonacci(2)
````

````python
def fibonacci_with_loop(n):
  a = 0
  b = 1
  for i in range(n):
    c = a + b
    a, b = b, c
  return c
````

````python
from math import sqrt

def fibonacci_analytical(n):
    return ((1+sqrt(5))**n - (1-sqrt(5))**n) / (2**n*sqrt(5))
````

````python
import time

start = time.time()
fibonacci(40)
end = time.time()
print('Recursive version:',round(end - start, 2))

start = time.time()
fibonacci_with_loop(40)
end = time.time()
print(f'Loop version: {end - start:.5f}')

start = time.time()
fibonacci_analytical(40)
end = time.time()
print('Analytical version: {:.5f}'.format(end - start))
````

##Real-World Cases Where Binary Search Uses Recursion

Searching in a Sorted Database of IDs

Hospitals, banks, and schools often store sorted lists of:

* student IDs

* patient records

* customer account numbers

* Navigation systems/GPS coordinate.

Binary search helps locate a record instantly, even if the database contains millions of entries.

###Recursive Binary Search

````python
def binary_search(arr, target, low, high):
    if low > high:
        return -1   # if target not found

    mid = (low + high) // 2

    if arr[mid] == target:
        return mid

    elif target < arr[mid]:
        return binary_search(arr, target, low, mid - 1)

    else:
        return binary_search(arr, target, mid + 1, high)

numbers = [2, 5, 7, 12, 15, 21, 30]
target = 21

result = binary_search(numbers, target, 0, len(numbers) - 1)

print("Target found at index:", result)
````

Binary search is naturally a divide-and-conquer process.
Every time you:

* Look at the middle.

* Decide left or right.

* Repeat the same process on a smaller part of the list.

Recursion expresses this repeating pattern clearly and cleanly.
