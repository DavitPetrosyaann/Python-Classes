#Dictionaries

Along with lists, dictionaries are one of the most flexible built-in object types in Python. If you think of
lists as order-based collections of objects, you can think of dictionaries as key-based collections; the
chief distinction is that in dictionaries, items are stored and fetched by key, instead of by positional
offset (indexes list[index_1]).

For example, dictionaries can replace many of the searching algorithms and data structures you might
have to implement manually in lower-level languages—as a highly optimized built-in tool, indexing a
dictionary is a very fast search operation. Dictionaries also sometimes do the work of records and
symbol tables used in other languages, can be used to represent sparse (mostly empty) data structures,
and much more. As a rundown of their main properties, Python dictionaries are:

- Accessed by key, not offset position
- Insertion-ordered collections of arbitrary objects
- Variable-length, heterogeneous, and arbitrarily nestable
- Of the category “mutable mapping”
- Tables of object references (hash tables)

##Changing Dictionaries in Place

Let’s continue with our interactive session. **Dictionaries, like lists, are mutable,** so you can change,
expand, and shrink them in place without making new dictionaries: simply assign a value to a key to
change or create an entry. The del statement works here, too; it deletes the entry associated with the key
specified as an index.

```python
D = {'hack': 1, 'Py': 2, 'code': 3}
print('Original dictionary: ', D, end = '\n\n')

D['Py'] = ['app', 'dev'] # Change entry (value=list)
print('Changed version: ', D, end = '\n\n')

del D['code'] # Delete entry (also D.pop(k))
print('One key-value pair is deleted: ', D, end = '\n\n')

D['years'] = 32 # Add new entry
print('One key-value pair is added: ', D, end = '\n\n')


D['Py'][0] = 'program' # Change a nested list in place
print('We changed the nested list item: ', D)
```

Like lists, assigning to an existing index in a dictionary changes its associated value. Unlike lists,
however, whenever you assign a new dictionary key (one that isn’t already present) you create a new
entry in the dictionary, as was done in the previous example for the key 'years'. This doesn’t work for
lists because you can only assign to existing list offsets—Python considers an offset beyond the end of a
list out of bounds and raises an error. As you learned earlier, to expand a list, you need to use tools such
as the append method or slice assignment instead.

```python
D = dict(a=1, b=2) # Literal keys stored left-to-right
D['c'] = 3 # New keys always added at the end
print('New key-value is added to dictionary: ', D, end = '\n\n')

D.pop('b') # Method pop removes any key by name
print('Other keys still in insertion order: ', D, end = '\n\n')

D['b'] = 2 # Earlier key goes at the end too
print('Earlier key goes at the end too: ', D, end = '\n\n')

D['c'] = 0 # Changing values doesn't impact order
print('Changing values doesn\'t impact order: ', D, end = '\n\n')

D['d'] = 1 # And new arrivals always go at the end
print('And new arrivals always go at the end: ', D, end = '\n\n')

D.popitem() #Getting item from Dictionary
print('Remove the last key-value pair from dictionary: ', D, end = '\n\n')

```

##Dictionary “Union” Operator

Per the quotes, though, this operation isn’t really mathematical set union at all. It’s a positionally
sensitive, key-based merge of dictionaries, that works the same as the dictionary update method but
returns its result as a new dictionary instead of updating a subject in place. Hence, dictionary | is like
the combination of running the dictionary’s copy and update methods in turn, and really just a minor
convenience for narrow roles. Here it is at work:

```python
D = dict(a=1, b=2)
print('Original dictionary: ', D, end = '\n\n')

B =  D | {'b': 3, 'c': 4} # Update, return, rightmost wins
print('Union of two dictionaires : ', B, end = '\n\n')

C = D | {'b': 3, 'c': 4} | dict(a=5, d=6)
print('Union of more than two dictionaires : ', C, end = '\n\n')
```

You get roughly the same mileage with the dictionary’s longstanding update method, though it requires a
manual copy to avoid in-place changes in programs that don’t want them (as a spoiler, the |= in-place
assignment form of dictionary union that you’ll meet later in this book is identical to calling update with
a single argument sans copy):

```python
print('Original dictionary: ', D, end = '\n\n')

C = D.copy()

print('Copy of original dictionary: ', C, end = '\n\n')

print('Are they the same object or not? ', id(D) == id(C), end = '\n\n')

C.update({'b': 3, 'c': 4}) # "Union" sans the | expression
print('Union operation 2nd way: ', C, end = '\n\n')

C.update(dict(a=5, d=6))
print('Union operation 3rd way: ', C, end = '\n\n')

# The Python 3.5 dictionary-literal unpacking we explored earlier
# can have the same effect too, though, at least when used with
# gnarly embedded literals as in the following, perhaps less readably:

E = {**D, **{'b': 3, 'c': 4}, **dict(a=5, d=6)} # ** iterable unpacking

print('Using ** operator for unpacking dictionaries : ', E, end = '\n\n')
```

##Using dictionaries for sparse data structures: Tuple keys


In a similar way, dictionary keys are also commonly leveraged to implement sparse—mostly empty—
data structures, such as multidimensional arrays where only a few positions have values stored in them:


```python
Matrix = {}             # Making dictionary
Matrix[(2, 3, 4)] = 88  # Creating new key-value in this case key is a tuple
Matrix[(7, 8, 9)] = 99  # Creating new key-value in this case key is a tuple

X = 2; Y = 3; Z = 4 # A ; separates statements

print('Getting the value by key: ', Matrix[(X, Y, Z)], end = '\n\n')
print('Getting the value by key: ', Matrix[(7, 8, 9)], end = '\n\n')

print('Just cheking the dictionary we created: ', Matrix)
```

## Avoiding missing-key errors

Errors for nonexistent key fetches like the foregoing are common in sparse matrixes, but you probably
won’t want them to shut down your program. There are at least three ways to fill in a default value
instead of getting such an error message

```python
print('We want to get dictionary item: ', Matrix.get((2, 3, 4), 0), end = '\n\n') # Exists: fetch and return

print('Getting value wich is not in dictionary and avoiding error: ', Matrix.get((2, 3, 6), 'No item')) # Doesn't exist: use passed default
```

# Tuples

The tuple object (pronounced “toople” or “tuhple,” depending on whom you ask) is
roughly like a list that cannot be changed—tuples are sequences, like lists, but they are
immutable, like strings. Functionally, they’re used to represent fixed collections of
items: the components of a specific calendar date, for instance. Syntactically, they are
normally coded in parentheses instead of square brackets and support arbitrary object
types, arbitrary nesting, and the usual sequence operations that we used on strings and
lists earlier:


```python
T = (1, 2, 3, 4) # 4-item tuple
t = 1,58,38, 'str'

print(type(T), T)
print(type(t), t)
```

Ordered collections of arbitrary objects
* Like strings and lists, tuples are positionally ordered collections of
  objects (i.e., they maintain a left-to-right order among their contents).
  Like lists and dictionaries, they can embed any kind of object.

Accessed by offset
* Like strings and lists, items in a tuple are accessed by offset (not by
key); they support all the offset-based access operations, such as
indexing and slicing.

Of the category “immutable sequence”
* Like strings and lists, tuples are sequences; they support many of the
same operations. However, like strings, tuples are immutable; they
don’t support any of the in-place change operations applied to lists.

Fixed-length, heterogeneous, and arbitrarily nestable
* Because tuples are immutable, you cannot change the size of a tuple
without making a copy. On the other hand, tuples may contain any
type of object, including other collection objects (e.g., lists,
dictionaries, and other tuples), and so support arbitrary nesting.

Arrays of object references
* Like lists, tuples are best thought of as object reference arrays: tuples
store access points to other objects (references), and indexing a tuple
is relatively quick.


```python
a = (4,)
type(a)
```

##Conversions, methods, and immutability

If you want
to sort a tuple, for example, you’ll usually have to either first **convert it to** a list to gain
access to a sorting method call and make it **a mutable object**, or use the newer sorted
built-in that accepts any sequence object.

```python
T = ('cc', 'aa', 'dd', 'bb')
tmp = list(T) # Make a list from a tuple's items
tmp.sort() # Sort the list
print('Sorted list from tuple: ', tmp, end = '\n\n')

T = tuple(tmp) # Make a tuple from the list's items
print('Tuple from list: ', T, end = '\n\n')

# Sorted(tuple) creates new list from tuple, because tuple immutable
print('Sorted List from Tuple: ', sorted(T, reverse=True), end = '\n\n') # Or use the sorted built-in, and save steps

print('Original tuple is not changed : ', T)
```

Although tuples don’t have the same methods as lists and strings, they do have two of
their **own—index and count** work as they do for lists, but they are defined for tuple
objects:

```python
T = (1, 2, 3, 2, 754, 74, 2) # Tuple methods
print('Offset of first appearance of 2: index _of_!: ', T.index(2), end = '\n\n')

print('Offset of appearance after offset 4: ', T.index(2, 4), end = '\n\n')

print('How many 798654s there are: ', T.count(798654))
```

Also, note that the rule about tuple immutability applies only to the top level of the
tuple itself, not to its contents. A list inside a tuple, for instance, can be changed as
usual:

```python
T = (1, [2, 3], 4)

T[1] = 'mod' # This fails: can't change tuple itself
```

```python
T[1][0] = 'mod' # This works: can change mutables inside
T
```

Like lists and dictionaries (we'll cover this later), tuples support mixed types and nesting, but they don’t grow
and shrink like lists and dictionaries because they are immutable:


```python
T = 'hack', 3.0, [11, 22, 33]

print('Original tuple T: ', T, end = '\n\n')

print('1st item of tuple T: ', T[1], end = '\n\n')

print('Get acces of nested items of tuple T: ', T[2][1], end = '\n\n')

T.append(4) # this method doesn't work on tuples
```

##Records Revisited: Named Tuples

We saw how to represent recordlike information with both a list and a dictionary and noted that dictionaries offer the
advantage of more mnemonic keys that label data. As long as we don’t require
mutability, tuples can serve similar roles, with positions for record fields like lists:


```python
pat = ('Pat', 40.5, ['dev', 'mgr'])
pat[1]
```

As for lists, though, field numbers in tuples generally carry less information than the
names of keys in a dictionary. To review, here’s the same record recoded as a
dictionary with named fields:


```python
pat = dict(name='Pat', age=40.5, jobs=['dev', 'mgr']) # Dictionary record

print('Original Dictionary: ', pat, end = '\n\n')

pat['name'], pat['jobs'] # Access by key, here we got tuple
```

In fact, we can convert parts of the dictionary to tuples if needed:

```python
print('Dictionary values to tuple: ', tuple(pat.values()), end = '\n\n')

print('Dictionary items to tuple list: ', list(pat.items()), end = '\n\n')
```

But really, this is a false dichotomy: with extra code, we can implement objects that
offer both positional and named access to record fields. For example, the namedtuple
utility.

```python
from collections import namedtuple                   # Import extension type
Rec = namedtuple('Rec', ['name', 'age', 'jobs'])     # Make a generated class
pat = Rec('Pat', age = 40.5, jobs = ['dev', 'mgr'])  # A named-tuple record
print(type(pat), end = '\n\n')
print(pat, end = '\n\n')

print('Access by position: ', pat[0], pat[2], end = '\n\n')
print('Access by attribute: ', pat.name, pat.jobs)
```

Converting to a dictionary also supports key-based behavior when needed:

```python
D = pat._asdict() # Dictionary-like form
print('Dictionary from named-tuple: ', D, end = '\n\n')

print('Access by key too: ', D['name'], D['jobs'])
```

As you can see, named tuples are a tuple/class/dictionary hybrid. They also represent a
classic trade-off. In exchange for their extra utility, they require extra code to use (the
two startup lines in the preceding examples that import the type and make the class) and
incur some performance costs to work this magic. Still, they are an example of the kind
of custom data types that we can build on top of built-in types like tuples when extra
utility is desired. They are also extensions, not core types—they live in the standard
library and fall into the same category as Chapter 5’s Fraction and Decimal—so
we’ll delegate to the Python library manual for more details.


##Definition:

A named tuple is a subclass of the regular tuple that allows you to access elements by name instead of just by index, while still being immutable.

Regular tuples: you access elements by position: t[0], t[1], …

Named tuples: you access elements by attribute names: t.name, t.age, …
