#Sets

Python sets are neither mappings nor sequences; rather, they are unordered collections
of immutable (technically, “hashable”) objects, which store each object just once. You
create sets by calling the built-in set function with a sequence or other iterable, or by
using a set literal expression, and sets support the usual mathematical set operations:


```python
x = set('Դու մի՛ հավատա ժըպտուն աչքերին') # Sequence => set

y = {'a', 'b', 1, 'c', 78, 0}                       # Set literal

print(x, end = '\n\n')
print(y)
```

Usual mathematical set operations

```python
x, y = {2, 1, 3}, {1, 2, 4}

print('Intersection of two sets: ', x & y, end = '\n\n')          # Intersection
print('Union of two sets: ', x | y, end = '\n\n')                 # Union
print('Difference of two sets: ', x - y, end = '\n\n')            # Difference
print('Superset of two sets: ', x > y, end = '\n\n')              # Superset
print('Symmetric difference of two sets: ', x ^ y, end = '\n\n')  # Symmetric difference
```

Sets can only contain hashable (immutable) types:

✅ Allowed: int, float, str, tuple, bool

❌ Not allowed: list, dict, set

# All hashable objects are immutable, but not all immutable objects are hashable.

hat does “immutable” mean?

An object is immutable if its contents cannot be changed after creation.

Examples:

✅ Immutable: `int`, `float`, `str`, `tuple`

❌ Mutable: `list`, `set`, `dict`

##Relationship between Hashable and Immutable:

All mutable objects are unhashable, because if you could change them, their hash value would also change — breaking `set`/`dict` behavior.

Most immutable objects are hashable — e.g. `int`, `str`, `tuple`.

But there’s an **exception**: if a `tuple` contains a mutable element, the whole `tuple` becomes unhashable!

Example:

```python
hash('RAFgnhmjhgOjhi')
```

```python
print(hash((1, 2, 3)), end = '\n\n')     # ✅ Works — all elements are immutable
#print(hash((1, [2, 3])))                 # ❌ Error — contains a mutable list
```

In Python, immutable objects are those whose value cannot be changed after creation — for example, numbers, strings, and tuples.
An object is hashable if it has a fixed hash value and can be used as a key in a dictionary or as an element in a set.

All mutable objects (like lists, sets, and dictionaries) are unhashable, because if their contents could change, their hash value would change too — breaking how sets and dicts work.

Most immutable objects are hashable, but not all: for example, a tuple is hashable only if all its elements are immutable.

Even less mathematically inclined programmers often find sets useful for common tasks

## How to offset item from `set`?

`Sets` are unordered, so they don’t support indexing or positional access.
That means you can’t “offset” (shift by position) like you can in `lists` or `tuples`.

```python
a = {1, 27, 34, 758}
a[1]           # set object is not subscriptable
```

But here’s what you can do instead:

Remove a specific item

```python
a = {1, 27, 34, 758}

a.discard(50)         # Removes 50 if found; does nothing otherwise
print('Original Set : ', a, end = '\n\n')

a.remove(27)          # Removes 27; raises error if not found
print('Set without 27: ', a)
```

Pop (remove and return) a random item

```python
a = {27, 34, 758, 1, 4, 55, 741, 78, 78}

x = a.pop()
print("Removed:", x)
print("Remaining set:", a)
```

`pop()` removes and returns an arbitrary element — not “the leftmost” or “the first.”
If you need a predictable order, use a list instead of a set.

##Common Set Methods

| **Method**                             | **Description**                                                         |                                                                      |
| -------------------------------------- | ----------------------------------------------------------------------- | -------------------------------------------------------------------- |
| `S.add(x)`                             | Adds element `x` to the set `S`.                                        |                                                                      |
| `S.remove(x)`                          | Removes element `x` from the set. Raises an error if `x` is not found.  |                                                                      |
| `S.discard(x)`                         | Removes element `x` if present; does **not** raise an error if missing. |                                                                      |
| `S.pop()`                              | Removes and returns an **arbitrary** element from the set.              |                                                                      |
| `S.clear()`                            | Removes **all** elements from the set, making it empty.                 |                                                                      |
| `S.union(A)` or `S                     | A`                                                                      Returns a new set containing **all unique elements** from both sets. |
| `S.intersection(A)` or `S & A`         | Returns elements **common** to both sets.                               |                                                                      |
| `S.difference(A)` or `S - A`           | Returns elements in `S` but **not** in `A`.                             |                                                                      |
| `S.symmetric_difference(A)` or `S ^ A` | Returns elements that are in `S` or `A` but **not in both**.            |                                                                      |
| `S.issubset(A)`                        | Returns `True` if all elements of `S` are in `A` (`S ⊂ A`).             |                                                                      |
| `S.issuperset(A)`                      | Returns `True` if all elements of `A` are in `S` (`A ⊂ S`).             |                                                                      |
| `S.isdisjoint(A)`                      | Returns `True` if `S` and `A` have **no common elements**.              |                                                                      |

#Frozen Set

A `frozenset` is just like a normal `set`, except it is **immutable** —
meaning once it’s created, you can’t add, remove, or modify its elements.

Because of this immutability, `frozensets` are hashable, so they can be:

* Used as `dictionary` keys, or

* Stored inside other `sets` (which normal sets cannot do).

##Creating a Frozenset

```python
# Create a frozenset from a list or set
fs = frozenset([1, 2, 3, 4])
print(fs)
```

## Frozenset Methods

| **Method**                       | **Description**                                                        |
| -------------------------------- | ---------------------------------------------------------------------- |
| `fs.copy()`                      | Returns a shallow copy of the frozenset.                               |
| `fs.union(other)`                | Returns a new frozenset containing all unique elements from both sets. |
| `fs.intersection(other)`         | Returns a frozenset containing elements common to both sets.           |
| `fs.difference(other)`           | Returns a frozenset with elements in `fs` but not in `other`.          |
| `fs.symmetric_difference(other)` | Returns elements in either `fs` or `other`, but not both.              |
| `fs.issubset(other)`             | Returns `True` if all elements of `fs` are in `other`.                 |
| `fs.issuperset(other)`           | Returns `True` if all elements of `other` are in `fs`.                 |
| `fs.isdisjoint(other)`           | Returns `True` if `fs` and `other` have no elements in common.         |


##Not Allowed (Mutating Methods)

Because `frozensets` are immutable, they do not support:

`add()`

`remove()`

`discard()`

`pop()`

`clear()`

#Files

You may already be familiar with the notion of files, which are named storage
compartments on your PC, phone, or other computer that are managed by your operating
system. The last major built-in object type that we’ll examine on our object-types tour
provides a way to access those files inside Python programs.

In short, the built-in open function creates a Python file object, which serves as a link to
a file residing on your device. After calling open, you can transfer strings of data to and
from the associated external file by calling the returned file object’s methods.

Compared to the types you’ve seen so far, file objects are outliers. They are considered
a core type because they are created by a built-in function, but they’re not numbers,
sequences, or mappings, and they don’t respond to expression operators; they export
only methods for common file-processing tasks. Most file methods are concerned with
performing input from and output to the external file associated with a file object, but
other file methods allow us to seek to a new position in the file, flush output buffers, and
so on. The following table summarizes common file operations.

##File operations methods

| **Operation**                        | **Interpretation / Description**                                    |
| ------------------------------------ | ------------------------------------------------------------------- |
| `output = open(r'C:\\data', 'w')`    | Create output file (Windows path, `'w'` = write).                   |
| `input = open('/home/me/data', 'r')` | Create input file (Unix path, `'r'` = read).                        |
| `input = open('data')`               | Create input file (current directory, `'r'` is default).            |
| `aString = input.read()`             | Read the entire file into a single string.                          |
| `aString = input.read(N)`            | Read up to the next **N characters** (or bytes) into a string.      |
| `aString = input.readline()`         | Read the next line (including `\n`) into a string.                  |
| `aList = input.readlines()`          | Read the entire file into a **list of lines** (with `\n`).          |
| `output.write(aString)`              | Write a string of characters (or bytes) into a file.                |
| `output.writelines(aList)`           | Write all lines from a list into a file **verbatim**.               |
| `output.close()`                     | Manually close the file (done automatically on garbage collection). |
| `output.flush()`                     | Flush the output buffer to disk **without closing** the file.       |
| `anyFile.seek(N)`                    | Move file pointer to offset **N** for next read/write operation.    |
| `for line in open('data'):`          | File iterator that reads line by line (memory efficient).           |
| `open('f.txt', encoding='utf-8')`    | Open a **Unicode text file** (using `str` strings).                 |
| `open('f.bin', 'rb')`                | Open a **binary file** (using `bytes` strings).                     |
| `codecs.open('f.txt', ...)`          | Alternative interface for Unicode text files.                       |


##Opening Files

To open a file, a program calls the built-in `open` function, with the external filename
first, followed by a processing mode. Both arguments are strings. The call returns a file
object, which in turn has methods for data transfer:

    afile = open(filename, mode)
    afile.method()

The first argument to `open`, the external filename, may include a platform-specific and
absolute (complete) or relative (partial) directory-path prefix that identifies a file’s
location in the host device’s filesystem. The filesystem is just a hierarchy of folders that
stores files and nested folders. A filename a/b/file.txt, for example, includes the path
prefix a/b that leads to a file on Unix (e.g., macOS, Linux, or Android), and a\b\file.txt
does the same on Windows.


The second argument to **open** is the processing `mode`, with options such as:
* `r` (default) to open text input  
* `w` to create and open for text output
* `a` to open for appending text to the end (for example adding logfiles).

## Storing Native Python Objects: pickle
Using `eval` to convert from strings to objects, as demonstrated in the above code, is a powerful tool. `eval` will run any Python expression - even one that might delete all the files on your computer, given the necessary permissions! If you really want to store native Python objects, but you can’t trust the source of the data in the file, Python’s standard library pickle module is ideal.

The `pickle` module is a more advanced tool that allows us to store almost any Python object in a file directly, with no to- or from-string conversion requirement on our part.

```python
# let's use the same my_dict dictionary
# and store it directly on the pickle file
pickle_file = open('datafile.pkl', 'wb')
# 'wb' mode means 'write binary'
```

```python
# we need to import the library for dealing with pickle files
import pickle
pickle.dump(my_dict, pickle_file)
pickle_file.close()
```

```python
# let's get back to the stored dictionary
pickle_file = open('datafile.pkl', 'rb')  # 'rb' mode means 'read binary'
same_dict = pickle.load(pickle_file)
same_dict
```

```python
type(same_dict)
```

`pickle` module performs all the object to and from byte string conversions for us. We can see how the byte strings look like.

```python
# it's good that we don't have to deal with the conversion
open('datafile.pkl', 'rb').read()
```

## Storing Python Objects in JSON Format

JSON is a newer and emerging data interchange format. It does not support as broad a range of Python object types as pickle, but its portability is an advantage in some contexts, and it represents another way to serialize
a specific category of Python objects for storage and transmission. Moreover, because JSON is so close to Python dictionaries and lists in syntax, the translation to and from Python objects is trivial, and is automated by the **json** standard library module.

```python
# json module translates
# the Python objects (for example the dict)
# to and from a JSON serialized string representation
# in memory
import json
json.dumps(my_dict)
```

We can easily translate Python objects to and from JSON data strings
in files. Before being stored in a file, your data is simply Python objects; the JSON module recreates them from the JSON textual representation when it loads it from the file

```python
json.dump(my_dict, fp=open('json_file.txt', 'w'), indent=4)
print(open('json_file.txt').read())
```

```python
# re-create the dict
same_dict = json.load(open('json_file.txt'))
same_dict
```

```python
some_list = json.load(open('sample_data/anscombe.json'))
some_list
```
